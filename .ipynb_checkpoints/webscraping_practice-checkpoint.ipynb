## Web Scraping in Python

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.scrapethissite.com/pages/forms/"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, "html.parser")
```

```python
print(soup.prettify())
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.scrapethissite.com/pages/forms/'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('div')
```

```python
soup.find_all('div')
```

```python
soup.find_all('div', class_ = 'col-md-12')
```

```python
soup.find_all('p')
```

```python
soup.find_all('p', class_ = 'lead')
```

```python
soup.find('p', class_ = 'lead').text.strip()
```

```python
soup.find_all('th')
```

```python
soup.find('th').text.strip()
```

```python
soup.find_all('td')
```

## Scraping Data from a Real Website

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_largest_law_firms_by_revenue"
```

```python
page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('table')
```
